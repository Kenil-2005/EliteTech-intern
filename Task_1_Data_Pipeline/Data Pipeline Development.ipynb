This project performs an automated ETL pipeline on Netflix TV shows and Movies dataset. It includes extraction, transformation (cleaning, encoding), and loading using Python.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv('netflix_titles.csv')

# Preview the dataset
df.head()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)

# Handle missing values

df['country'].fillna('Unknown', inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['rating'].fillna('Not Rated', inplace=True)

# Feature engineering
df['year_bin'] = pd.cut(df['release_year'], bins=[1920, 1980, 2000, 2010, 2020, 2025], labels=['1920s-80', '80-00', '00-10', '10-20', '20+'])

# Encode type column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['type_encoded'] = le.fit_transform(df['type'])
# Create year bins and assign string labels
bins = [1920, 1980, 2000, 2010, 2020, 2025]
labels = ['1920-1980', '1981-2000', '2001-2010', '2011-2020', '2021-2025']

df['year_bin'] = pd.cut(df['release_year'], bins=bins, labels=labels)
```
